```python
import scipy #used for optimize
import numpy as np
from scipy.special import expit
import scipy.io
import random
import matplotlib.pyplot as plt
import matplotlib.cm 	as cm
```

### Sigmoid Function

```python
def sigmoidFunc(X, theta):
    return expit(np.dot(X.astype(np.float), theta.astype(np.float)))
```

#### Give a vector, predic the number based on index with max probability

```python
def predictnumber(hyp_prob):
    numbers = range(1,11)
    predicted_number = numbers[np.argmax(np.array(hyp_prob))]
    return predicted_number
```
