```python
import scipy #used for optimize function
import numpy as np
from scipy.special import expit #sigmoid function
import scipy.io #accept mat input
from scipy import optimize
import random
import matplotlib.pyplot as plt
import matplotlib.cm as cm
```

##### Sigmoid Function

```python
def sigmoidFunc(X, theta):
    return expit(np.dot(X.astype(np.float), theta.astype(np.float)))
```

##### Gradient for Logistic Regression.

```python
def gradientFunc(theta, X, y, lamda=0.):
    theta = np.asmatrix(theta)
    theta = theta.T
    h_x = sigmoidFunc(X, theta)
    h_x_y = np.subtract(h_x,y)
    cols = X.shape[1]
    gradient = np.zeros((cols,1))
    gradient = (np.dot(h_x_y.T, X) * (1./m))

    lamda_term = (lamda/m) * theta
    #except for the first term
    lamda_term[0,0] = 0;
    gradient = gradient.T +lamda_term

    gradient = np.squeeze(np.asarray(gradient))
    return gradient
```

### Cost Function for Logistic Regression

```python
def costFunc(theta, X, y, lamda=0.):
    theta = np.asmatrix(theta)
    theta = theta.T
    h_x = sigmoidFunc(X, theta)
    log_h_x = np.log(h_x)
    lhs = np.multiply(y, log_h_x)
    rhs = np.multiply((1 - y), np.log(1-h_x))
    cost = - lhs - rhs# np.dot(minus_y_t, log_minus_h_x)))
    cost = np.sum(cost)*(1./m)

    #lamda term for the regularization
    lamda_term = np.square(theta)*(lamda/(2*m))
    #except for the first term
    lamda_term[0,0] = 0
    lamda_term = np.sum(lamda_term)
    cost = cost + lamda_term
    return cost
```

```python
#get the square matrix from the row
def getDatumImg(row):
    width, height = 20,20
    square = row[1:].reshape(width, height)
    return square.T

#Display the 100 random images or the specified indices
def displayData(indices_to_display = None):
    width, height = 20, 20
    nrows, ncols = 10,10
    if not indices_to_display:
        indices_to_display = random.sample(range(X.shape[0]), nrows*ncols)
    big_pic = np.zeros((height*nrows, width*ncols))
    irow, icol = 0, 0
    for idx in indices_to_display:
        if icol == ncols:
            irow += 1
            icol = 0
        iimg = getDatumImg(X[idx])
        big_pic[irow*height:irow*height+iimg.shape[0], \
                icol*width:icol*width+iimg.shape[1]] = iimg
        icol += 1

    fig = plt.figure(figsize=(6,6))
    img = scipy.misc.toimage(big_pic)
    plt.imshow(img, cmap = cm.Greys_r)
    plt.show()
```

### Function predicts the number on the training data using calculated optimum Theta. Selects the maximum value Out of the all predicted probability.

```python
def predictOneAll(opt_theta, row):
    numbers = [0] + range(1,10)
    hyp_prob = [0] * len(numbers)
    for i in range(10):
        hyp_prob[i]= sigmoidFunc(X[row,:], opt_theta[i,:].T)
    #choose the maximum of the predictions
    predicted_number = numbers[np.argmax(np.array(hyp_prob))]
    return predicted_number
```

### calculates the accuracy of prediction on the test data and returns the list of inaccurate predictions.

```python
def calculateAccuracy(opt_theta, X, y):
    totImages = X.shape[0]
    corrects = 0.
    inaccurate = []
    for i in range(totImages):
        predicted_value = predictOneAll(opt_theta, i)
        if (predicted_value+1) == y[i,0]:
            corrects += 1
        else:
            inaccurate.append(i)
    print "total number of corrects = ", corrects
    print "The accuracy of the algorithm = ", corrects/X.shape[0]
    return inaccurate
```

```python
n = X.shape[1]
opt_theta= np.zeros((10,n))
lamda = 0.
for i in range(10):
    new_y = np.copy(y)
    new_y[new_y != i+1] = 0
    new_y[new_y !=0] = 1
    theta = np.zeros((1,n))
    print "calculating optimal theta for the images: ", i
    #you can try other otimize functions such as BFGS, TNC, Newton-CG
    result = optimize.minimize(fun=costFunc, x0=theta, args=(X,new_y,lamda),
                               method = 'CG',
                               options={'maxiter':50, 'disp':False},
                               jac=gradientFunc)
    #result = optimize.fmin_cg(costFunc, fprime=gradientFunc, x0= theta, \
                          #args = (X,new_y), maxiter = 50)
    opt_theta[i,:] = np.asmatrix(result['x'].T)#['x']
```

```python
inacurate = calculateAccuracy(opt_theta,X,y)
```
